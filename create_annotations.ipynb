```python
src_dir = "/home/michael/bee_cam/data/datasets/class10_allclass/train"
dest_dir = "/home/michael/ct_classifier/data/allclass/train"
image_json = []
annot_json = []

if not os.path.exists(dest_dir):
    os.makedirs(dest_dir)

image_id = 0
annotation_id = 0

for file in os.listdir(src_dir):
    if file.endswith(".txt"):
        base_name = os.path.splitext(file)[0]
        image_path = os.path.join(src_dir, base_name + ".jpg")
        try:
            with Image.open(image_path) as img:
                width, height = img.size
                with open(os.path.join(src_dir, file), 'r') as f:
                    for line in f.readlines():
                        category_id, x, y, w, h = map(float, line.strip().split())
                        category_id = int(category_id)
                        
                        # Convert xywh to x1, y1, x2, y2 format
                        x1 = int((x - w/2) * width)
                        y1 = int((y - h/2) * height)
                        x2 = int((x + w/2) * width)
                        y2 = int((y + h/2) * height)
                        
                        cropped = img.crop((x1, y1, x2, y2))
                        cropped_file_name = f"{base_name}_{annotation_id}.jpg"
                        cropped_path = os.path.join(dest_dir, cropped_file_name)
                        cropped.save(cropped_path)
                        
                        cropped_width, cropped_height = cropped.size
                        area = cropped_width * cropped_height
                        
                        image_json.append({
                            'file_name': cropped_file_name,
                            'height': cropped_height,
                            'width': cropped_width,
                            'id': image_id
                        })
                        
                        annot_json.append({
                            'id': annotation_id,
                            'image_id': image_id,
                            'category_id': category_id,
                            'is_crowd': 0,
                            'segmentation': [],
                            'area': area
                        })
                        
                        image_id += 1
                        annotation_id += 1
        except:
            continue

# Save the JSON data
with open('data/allclass/train/images.json', 'w') as f:
    json.dump(image_json, f)

with open('data/allclass/train/annotations.json', 'w') as f:
    json.dump(annot_json, f)
```
